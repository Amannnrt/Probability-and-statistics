# Univariate Gaussian Distribution

The univariate Gaussian distribution, also known as the normal distribution, is a continuous probability distribution characterized by its mean $ \mu $ and variance $ \sigma^2 $ . The probability density function (PDF) of a univariate Gaussian random variable $ X \sim \mathcal{N}(\mu, \sigma^2) $ is given by:

$$
p(x) = \frac{1}{\sqrt{2 \pi \sigma^2}} \exp\left(-\frac{(x - \mu)^2}{2 \sigma^2}\right), \quad x \in \mathbb{R}.
$$

Here:

- $ \mu \in \mathbb{R} $ is the mean of the distribution, representing the center of the bell curve.
- $ \sigma^2 > 0 $ is the variance, representing the spread of the distribution.
- $ \sigma = \sqrt{\sigma^2} $ is the standard deviation.

The term $ (x - \mu)^2 $ measures the squared deviation of $ x $) from the mean $ \mu $, and the exponential term ensures that values closer to $ \mu $ have higher probabilities. The normalization constant $ \frac{1}{\sqrt{2 \pi \sigma^2}} $ ensures that the total probability integrates to 1.

---

# Bivariate Gaussian Distribution

The bivariate Gaussian distribution extends the univariate case to two dimensions. It describes the joint distribution of two random variables $ (X_1, X_2) $ that are jointly Gaussian. The PDF of a bivariate Gaussian distribution is given by:

$$
p(x_1, x_2) = \frac{1}{2 \pi \sqrt{|\Sigma|}} \exp\left(-\frac{1}{2} (x - \mu)^\top \Sigma^{-1} (x - \mu)\right),
$$

where:

- $ x = \begin{bmatrix} x_1 \\ x_2 \end{bmatrix} $ is the vector of random variables.
- $ \mu = \begin{bmatrix} \mu_1 \\ \mu_2 \end{bmatrix} $ is the vector of means.
- $ \Sigma = \begin{bmatrix} \sigma_1^2 & \rho \sigma_1 \sigma_2 \\ \rho \sigma_1 \sigma_2 & \sigma_2^2 \end{bmatrix} $ is the covariance matrix, -where:
 - $ \sigma_1^2 $  and  $\sigma_2^2 $ are the variances of X_1  and  X_2 , respectively.
 - $ \rho $ is the correlation coefficient between X_1  and X_2 , satisfying $ -1 \leq \rho \leq 1 $.
 - $ |\Sigma| $ is the determinant of the covariance matrix $ \Sigma $.
 - $ \Sigma^{-1} $ is the inverse of the covariance matrix.

---

## Expanded Form of the Bivariate Gaussian PDF

Let us expand the quadratic form in the exponent:

$$
(x - \mu)^\top \Sigma^{-1} (x - \mu),
$$

where $ x - \mu = \begin{bmatrix} x_1 - \mu_1 \\ x_2 - \mu_2 \end{bmatrix} $. Using the explicit form of $ \Sigma^{-1} $, we can write:

$$
\Sigma^{-1} = \frac{1}{|\Sigma|} \begin{bmatrix} \sigma_2^2 & -\rho \sigma_1 \sigma_2 \\ -\rho \sigma_1 \sigma_2 & \sigma_1^2 \end{bmatrix},
$$

where $ |\Sigma| = \sigma_1^2 \sigma_2^2 (1 - \rho^2) $.

Substituting this into the quadratic form:

$$
(x - \mu)^\top \Sigma^{-1} (x - \mu) = \frac{1}{1 - \rho^2} \left[ \frac{(x_1 - \mu_1)^2}{\sigma_1^2} - \frac{2 \rho (x_1 - \mu_1)(x_2 - \mu_2)}{\sigma_1 \sigma_2} + \frac{(x_2 - \mu_2)^2}{\sigma_2^2} \right].
$$

Thus, the bivariate Gaussian PDF becomes:

$$
p(x_1, x_2) = \frac{1}{2 \pi \sigma_1 \sigma_2 \sqrt{1 - \rho^2}} \exp\left(-\frac{1}{2 (1 - \rho^2)} \left[ \frac{(x_1 - \mu_1)^2}{\sigma_1^2} - \frac{2 \rho (x_1 - \mu_1)(x_2 - \mu_2)}{\sigma_1 \sigma_2} + \frac{(x_2 - \mu_2)^2}{\sigma_2^2} \right]\right).
$$

---

# Comparing Univariate and Bivariate Gaussians

The univariate and bivariate Gaussian distributions share the same fundamental structure but differ in dimensionality and notation. Let us compare their key components:

### Exponential Term:
- In the univariate case:
  $$
  \exp\left(-\frac{(x - \mu)^2}{2 \sigma^2}\right).
  $$
- In the bivariate case:
  $$
  \exp\left(-\frac{1}{2} (x - \mu)^\top \Sigma^{-1} (x - \mu)\right).
  $$

Here, $ \frac{(x - \mu)^2}{\sigma^2} $ generalizes to the quadratic form $ (x - \mu)^\top \Sigma^{-1} (x - \mu) $.

### Normalization Constant:
- In the univariate case:
  $$
  \frac{1}{\sqrt{2 \pi \sigma^2}}.
  $$
- In the bivariate case:
  $$
  \frac{1}{2 \pi \sqrt{|\Sigma|}}.
  $$

Here, $ \frac{1}{\sqrt{2 \pi \sigma^2}} $ generalizes to $ \frac{1}{2 \pi \sqrt{|\Sigma|}} $, where $ |\Sigma| $ is the determinant of the covariance matrix.

### Mean and Variance:
- In the univariate case, the distribution is parameterized by $ \mu $ and $ \sigma^2 $.
- In the bivariate case, the distribution is parameterized by $ \mu $ (a vector) and $ \Sigma $ (a matrix).

### Geometric Interpretation:
- In the univariate case, the distribution is symmetric about the mean \( \mu \) along the real line.
- In the bivariate case, the distribution forms an elliptical contour in the plane, with the orientation and shape determined by the covariance matrix \( \Sigma \).


### General Form of the Bivariate Gaussian PDF

The general form of the bivariate Gaussian probability density function (PDF) is:

$$
p(x_1, x_2) = \frac{1}{2 \pi \sigma_1 \sigma_2 \sqrt{1 - \rho^2}} \exp\left(- \frac{1}{2(1 - \rho^2)} \left[ \frac{(x_1 - \mu_1)^2}{\sigma_1^2} - \frac{2 \rho (x_1 - \mu_1)(x_2 - \mu_2)}{\sigma_1 \sigma_2} + \frac{(x_2 - \mu_2)^2}{\sigma_2^2} \right] \right).
$$

When $ \rho = 0 $, the covariance matrix $ \Sigma $ becomes diagonal:

$$
\Sigma = \begin{bmatrix} 
\sigma_1^2 & 0 \\ 
0 & \sigma_2^2 
\end{bmatrix}.
$$

This implies:

- The determinant of $ \Sigma $ is $ |\Sigma| = \sigma_1^2 \sigma_2^2 $.
- The inverse of $ \Sigma $ is:
  $
  \Sigma^{-1} = \begin{bmatrix} 
  \frac{1}{\sigma_1^2} & 0 \\ 
  0 & \frac{1}{\sigma_2^2} 
  \end{bmatrix}.
  $

Substituting $ \rho = 0 $ into the PDF, we get:

$$
p(x_1, x_2) = \frac{1}{2 \pi \sigma_1 \sigma_2} \exp\left(- \frac{1}{2} \left[ \frac{(x_1 - \mu_1)^2}{\sigma_1^2} + \frac{(x_2 - \mu_2)^2}{\sigma_2^2} \right] \right).
$$

Notice that the cross-term $ - \frac{2 \rho (x_1 - \mu_1)(x_2 - \mu_2)}{\sigma_1 \sigma_2} $ vanishes because $ \rho = 0 $.

---

### Factorization into Independent Univariate Gaussians

The above expression can now be factored as:

$$
p(x_1, x_2) = \left( \frac{1}{\sqrt{2 \pi \sigma_1^2}} \exp\left(- \frac{(x_1 - \mu_1)^2}{2 \sigma_1^2} \right) \right) \cdot \left( \frac{1}{\sqrt{2 \pi \sigma_2^2}} \exp\left(- \frac{(x_2 - \mu_2)^2}{2 \sigma_2^2} \right) \right).
$$

Here:

- The first term is the PDF of a univariate Gaussian for $ X_1 \sim \mathcal{N}(\mu_1, \sigma_1^2) $.
- The second term is the PDF of a univariate Gaussian for $ ( X_2 \sim \mathcal{N}(\mu_2, \sigma_2^2) $.

Thus, when \( \rho = 0 \), the joint distribution $ p(x_1, x_2) $ is the product of two independent univariate Gaussian distributions:

$$
p(x_1, x_2) = p(x_1) \cdot p(x_2),
$$

where:

- $ p(x_1) = \frac{1}{\sqrt{2 \pi \sigma_1^2}} \exp\left(- \frac{(x_1 - \mu_1)^2}{2 \sigma_1^2} \right) $,
- $ ( p(x_2) = \frac{1}{\sqrt{2 \pi \sigma_2^2}} \exp\left(- \frac{(x_2 - \mu_2)^2}{2 \sigma_2^2} \right) $.

---

### Geometric Interpretation

When $ \rho = 0 $, the level curves of the bivariate Gaussian (the contours of constant probability density) are circular or elliptical with axes aligned along the coordinate axes. This reflects the **independence** of $ X_1 $ and $ X_2 $: knowing the value of one variable provides no information about the other.

In contrast, when $ \rho \neq 0 $, the level curves become tilted ellipses, indicating **dependence** between X_1 and \X_2 . The tilt depends on the sign and magnitude of  $ \rho$ .


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Function to plot bivariate Gaussian
def plot_bivariate_gaussian(mean, cov, title):
    # Create grid for plotting
    x, y = np.mgrid[-3:3:.01, -3:3:.01]
    pos = np.dstack((x, y))
    
    # Define the bivariate Gaussian distribution
    rv = multivariate_normal(mean, cov)
    
    # Evaluate the PDF on the grid
    z = rv.pdf(pos)
    
    # Plot the contours
    plt.figure(figsize=(6, 6))
    plt.contourf(x, y, z, levels=50, cmap='viridis')
    plt.colorbar(label='Probability Density')
    plt.title(title)
    plt.xlabel('$x_1$')
    plt.ylabel('$x_2$')
    plt.grid(False)
    plt.axis('equal')
    plt.show()

# Case 1: Uncorrelated variables (rho = 0)
mean_uncorrelated = [0, 0]  # Mean vector
cov_uncorrelated = [[1, 0],  # Covariance matrix (diagonal)
                    [0, 1]]
plot_bivariate_gaussian(mean_uncorrelated, cov_uncorrelated, "Bivariate Gaussian (ρ = 0)")

# Case 2: Correlated variables (rho = 0.8)
mean_correlated = [0, 0]  # Mean vector
cov_correlated = [[1, 0.8],  # Covariance matrix (non-diagonal)
                  [0.8, 1]]
plot_bivariate_gaussian(mean_correlated, cov_correlated, "Bivariate Gaussian (ρ = 0.8)")
```
